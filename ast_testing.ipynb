```python
from sys import argv
import pygraphviz as pgv
import pandas as pd

# read CSV file (argument 1) and produce DOT+PNG (argument 2)

#df = pd.read_csv(argv[1])
G = pgv.AGraph(strict=False, directed=True)

# file name is a node
for f_name in set(df["file_name"].tolist()):
    G.add_node(f_name)

# function name is a node
for func_name in set(df["name"].tolist()):
    G.add_node(func_name)

#df = df.loc[:20]
X = df.itertuples()
while True:
    try:
        resp = next(X)[1:]
        G.add_edge(resp[0], resp[1], label=resp[2])
    except StopIteration:
        break

G.write('test' + ".dot")




# dot -Ksfdp -ooutput.svg -Tsvg test.dot
```
